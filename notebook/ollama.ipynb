```python
from langchain_ollama import ChatOllama
```

```python
llm = ChatOllama(
    model="gemma3:1b",
    temperature=0,
    # other params...
)
```

```python
messages = [
    (
        "system",
        "You are a helpful assistant that translates English to Sanskrit. Translate the user sentence.",
    ),
    ("human", "I love programming."),
]
ai_msg = llm.invoke(messages)
ai_msg
```

```python
print(ai_msg.content)
```
